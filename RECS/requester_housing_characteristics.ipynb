```python
base_url = "https://www.eia.gov/consumption/residential/data"
outdir = "outputs/housing_characteristics"
"https://www.eia.gov/consumption/residential/data/2005/c&e/xls/tableus1part1.xls"
"https://www.eia.gov/consumption/residential/data/2020/hc/xls/HC%201.1.xlsx"
for year in  [1993, 1997, 2001, 2005, 2009, 2015, 2020]:
   year_dir = os.path.join(outdir, str(year))
   if not os.path.exists(year_dir):
      os.mkdir(year_dir)
   curr_url = base_url + "/" + str(year)
   resp = requests.get(curr_url)
   if resp.status_code == 200:
      soup = BeautifulSoup(resp.content)
      links = soup.find_all('a', href = True)
      links = [x['href'] for x in links if '.xls' in x['href'] or ".pdf" in x['href']]
      print(links)

      for link in links:
         print("processing this link " + link)
         if "/consumption" in link:
            link = link[len("/consumption/residential/data/1993/"):]
         
         if link[0] != "/":
            full_link = curr_url + "/" + urllib.parse.quote(link)
         else:
            full_link = curr_url + urllib.parse.quote(link)
         print(full_link)
         link_resp = requests.get(full_link)
         table_name = re.findall("^.*\/(.*)$", link)[0]
         print(table_name)
         if (link_resp.status_code == 200):
            output = open(os.path.join(year_dir, table_name), 'wb')
            output.write(link_resp.content)
            output.close()
      
# https://www.eia.gov/consumption/residential/data/2020/hc/xls/HC%201.1.xlsx
# https://www.eia.gov/consumption/residential/data/1993/pdf/rx93htab.pdf
```
